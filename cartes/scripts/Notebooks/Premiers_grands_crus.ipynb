```python
# Open the file
with open('./jsons/data_bourgogne_wtype_wdes.json') as json_file:
    data = json.load(json_file)
    
file = open("./cartes/scripts/txt files/grandscrus.txt",'r')
grandscrus = file.read().split('\n')
for key in data.keys():
    for gc in grandscrus:
        if data[key]['properties']['appellation'] == gc :
            data[key]['properties']['Grand Cru'] = 1
```

```python
data_premier_rouge = {}
data_premier_blanc = {}
data_grand_rouge = {}
data_grand_blanc = {}

for key in data.keys():
    if data[key]['properties']['type_vin'] == 'rouge':
        if data[key]['properties']['Premier Cru']==1 :
            data_premier_rouge[key] = data[key]
        if data[key]['properties']['Grand Cru']==1 :
            data_grand_rouge[key] = data[key]
            
    if data[key]['properties']['type_vin'] == 'blanc':
        if data[key]['properties']['Premier Cru']==1 :
            data_premier_blanc[key] = data[key]
        if data[key]['properties']['Grand Cru']==1 :
            data_grand_blanc[key] = data[key]
```
